###Data Cleaning

```python
data['Reviews'].unique()
```

```python
data['Reviews'].str.isnumeric().sum()
```

```python
data[~data['Reviews'].str.isnumeric()]
```

```python
data_cpy=data.copy()
```

```python
data_cpy=data_cpy.drop(data_cpy.index[10472])
```

```python
data_cpy.head()
```

```python
data_cpy['Size'].unique()
```

```python
data_cpy.columns
```

```python
data_cpy['Size'] = data_cpy['Size'].str.replace('M', '000')
data_cpy['Size'] = data_cpy['Size'].str.replace('k', '')
data_cpy['Size'] = data_cpy['Size'].replace('Varies with device', np.nan)
data_cpy['Size'] = data_cpy['Size'].astype(float)
```

```python
data_cpy['Size'].unique()
```

```python
data_cpy.head()
```

```python
data_cpy['Installs'].unique()
```

```python
data_cpy['Price'].unique()
```

```python
remov = ['+',',','$']
remove_column = ['Installs', 'Price']

for col in remove_column:
    data_cpy[col] = data_cpy[col].astype(str)
    for char in remov:
        data_cpy[col] = data_cpy[col].str.replace(char, '', regex=False)
```

```python
data_cpy['Price'].isnull().sum()
```

```python
data_cpy['Installs'].isnull().sum()
```

```python
data_cpy['Installs'] = data_cpy['Installs'].astype(float)
data_cpy['Price'] = data_cpy['Price'].astype(float)
```

```python
data_cpy['Last Updated'] = pd.to_datetime(data_cpy['Last Updated'])
```

```python
data_cpy['day']=data_cpy['Last Updated'].dt.day
data_cpy['month']=data_cpy['Last Updated'].dt.month
data_cpy['year']=data_cpy['Last Updated'].dt.year
```
